```python
import pandas as pd
from statsmodels.tsa.stattools import grangercausalitytests

data = pd.read_csv("Dataset - Sheet1.csv")
test = pd.DataFrame(data)
print(test.head())
print(test.columns)

target = 'abundance in SNE area (millions of lobsters)'
for column in test.columns:
    if column != target:
        results = grangercausalitytests(test[[target, column]], maxlag=3, verbose=False)

        significant_lags = []
        for lag, result in results.items():
            p_value = result[0]['ssr_ftest'][1]
            if p_value < 0.05:
                significant_lags.append((lag, p_value))

        if significant_lags:
            print(f"\nGranger Causality Test: {column} -> {target}")
            for lag, p_value in significant_lags:
                print(f"Lag {lag}: Significant Granger Causality (p={p_value:.4f})")
```
